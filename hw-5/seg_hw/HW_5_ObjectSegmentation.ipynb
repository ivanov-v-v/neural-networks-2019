### Тут нужно нарисовать картинки, с результатими сегментации из тестового сета

```python
import matplotlib.pyplot as plt
from model import SegmenterModel
import torch
import numpy as np
import carvana_dataset as cv
%matplotlib inline

DEVICE_ID = 2
DEVICE = torch.device('cuda:%d' % DEVICE_ID)
torch.cuda.set_device(DEVICE_ID)
```

```python
train_data = cv.CarvanaDataset('data/train/', 'data/train_masks/')
test_data = cv.CarvanaDataset('data/test/', 'data/test_masks/')
```

```python
unet = SegmenterModel()
unet.load_state_dict(torch.load("unet_dump_recent"))
unet.eval();
```

```python
n_img = len(test_data.files)
img = test_data[np.random.randint(0, n_img)][0]
plt.imshow(img.permute(1, 2, 0));
```

```python
mask = unet.forward(img.unsqueeze(0).cuda())
plt.imshow((mask > 0).squeeze(0).squeeze(0).cpu());
```
